```python
dfTennis["TOURNAMENT_SURFACE"].value_counts()
#display the different surfaces played on 
#instead of having 8 rows it will be better just to have 3 rows 
```

```python
dfTennis["TOURNAMENT_SURFACE"].replace(to_replace=['DecoTurf - outdoors','outdoors','Rebound Ace','Plexicushion Prestige'],value='Hard Court',inplace=True)
```

```python
dfTennis["TOURNAMENT_SURFACE"].replace(to_replace=['Grass / Outdoor','Clay - outdoors'],value=['Grass','Clay'],inplace=True)
```

```python
dfTennis["TOURNAMENT_SURFACE"].value_counts()
#now the surface have been grouped into 3 rows 
```

```python
plt.figure(figsize=(8,9))
plt.pie(dfTennis["TOURNAMENT_SURFACE"].value_counts(),autopct='%.2f',frame=True,startangle=45,radius=2.5,colors=['green','steelblue','sandybrown'])
plt.legend(labels=['Grass','Hard Court','Clay'],loc=3)
plt.xticks(ticks=[])
plt.yticks(ticks=[])
plt.title("Number of matches on various types of courts")
#a ppie chhart to show the surfaces played on 
```

```python
plt.figure(figsize=(15,8))
counts,edges,barrs=plt.hist(dfTennis['TOURNAMENT'],bins=7,align='mid',label='Number of matches per Grand Slam recorded',color="green")
plt.title("Number of matches for each Grand Slam Tournament")
plt.xlabel("Grand Slams")
plt.ylabel("----- Number of recorded Matches ----->")
plt.bar_label(barrs)
plt.grid(axis='x')
plt.plot()
#barchart for the tournoments 
```

```python
plt.figure(figsize=(19,10))
sns.lineplot(data=dfTennis,x='YEAR',y='WINNER_PRIZE',hue='TOURNAMENT',linewidth=3.6,palette=['sandybrown','lightseagreen','cornflowerblue','green'])
plt.title("Rewards from each Grand Slam Tournament")
plt.xlabel("Years")
plt.ylabel(" Winner Prize($Million)")
plt.plot()
#graph represents the winnings in millions 
```

```python
dfTennis_pivot=pd.pivot_table(data=dfTennis,columns=['TOURNAMENT'],index='WINNER_NATIONALITY',values='YEAR',aggfunc='count')
ax=dfTennis_pivot.plot(kind='bar',figsize=(32,25),width=0.9,color=['lightseagreen','sandybrown','cornflowerblue','green'])
for c in ax.containers:
    ax.bar_label(c, fmt='%.0f', label_type='edge')
plt.title("Nationality of Grand Slam Tournament Winners")
plt.xlabel("Nationalities")
plt.ylabel("----- Number of Matches won ----->")
plt.xticks()
plt.yticks(ticks=[i for i in range(0,22,2)],labels= [i for i in range(0,22,2)])
plt.plot()
#created these two graphs to clearly dislay the winners, tournoments , playstyle and nationality and is much easier to read 
```

```python
dfTennis2_pivot=pd.pivot_table(data=dfTennis,columns=['TOURNAMENT'],index=['WINNER_NATIONALITY','WINNER_LEFT_OR_RIGHT_HANDED'],values='YEAR',aggfunc='count')
ax=dfTennis2_pivot.plot(kind='bar',figsize=(32,25),width=0.9,color=['lightseagreen','sandybrown','cornflowerblue','green'])
for c in ax.containers:
    ax.bar_label(c, fmt='%.0f', label_type='edge',padding=2)
plt.title("Distribution of Grand Slam Tournament Winners")
plt.xlabel("Nationalities and Playstyle")
plt.ylabel("----- Number of Matches won ----->")
plt.xticks(rotation=90)
plt.yticks(ticks=[i for i in range(0,22,2)],labels= [i for i in range(0,22,2)])
plt.plot()
```
